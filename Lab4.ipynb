### Import the IQ, height & weight, life expectancy, quality of life, and population density dataframes

```python
#DataFrames imported
iq_df = pd.read_csv('iq.csv')
height_weight_df = pd.read_csv('height_weight_data.csv')
expectancy_df = pd.read_csv('life_expectancy.csv')
quality_df = pd.read_csv('quality_of_life.csv')
populate_df = pd.read_csv('population_density.csv')
```

Cleaning Data

```python
#Getting Nans out of iq.education_expenditure...

iq_df['education_expenditure_per_inhabitant'].replace(',', '', regex=True,inplace = True)


#checking for NaNs:
iq_df[iq_df['education_expenditure_per_inhabitant'].isnull()]

#Getting rid of rows with NaN:
iq_df['education_expenditure_per_inhabitant'] = pd.to_numeric(iq_df['education_expenditure_per_inhabitant'], errors='coerce')
iq_df = iq_df.dropna(subset=['education_expenditure_per_inhabitant'])
iq_df['education_expenditure_per_inhabitant'] = iq_df['education_expenditure_per_inhabitant'].astype(int)

# Resetting the id
iq_df = iq_df.reset_index(drop=True)
iq_df.head(6)
```

4. Join the height and weight data to the iq data. 

```python
health_data = pd.merge( height_weight_df, iq_df, how = 'inner' )
health_data
```

5. Join population density to quality of life data.

```python
quality_df 
populate_df


#instead of doing how = 'right', I used 'inner' to drop NaN rows in the population dataframe.
population_quality = pd.merge( quality_df, populate_df, how = 'inner' )
population_quality
```

6. Join life expectancy to quality of life data.

```python
quality_df 
expectancy_df 

expect_quality = pd.merge( quality_df, expectancy_df, on = 'country' )
expect_quality
```

7. Join iq, life expectancy, quality of life, and population density into 1 dataframe.

```python
#Merge the first two DataFrames, then merge the others to it.

# health_data = height_weight and iq
# population_quality = population_density and quality_oflife
# expect_quality = life_expectancy and quality_oflife

# do not use (on = 'country') for this.
conditions = pd.merge(health_data, population_quality, how = 'inner')

one_dataframe = pd.merge(conditions, expect_quality, how = 'inner')
# one_dataframe.info() 
one_dataframe
```
